# How to get instagram data ?

**Libraries**

Instaloader : library python to extract instagram's data

*Instaloader 4.2.10*

```python
from instaloader import *
import os
import pathlib
import shutil
import lzma
import json
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
users = ["marty_the_cockerdale", "lecoindespatissiers", "cataniafoodprn", "guatemala_magica"]
```

Get Instagram posts data

**Instanciation of Instaloader object**

Documentation here : https://instaloader.github.io/as-module.html#instaloader-main-class

```python
L = instaloader.Instaloader()
users = ["marty_the_cockerdale", "lecoindespatissiers", "cataniafoodprn", "guatemala_magica", "tina.kandybina"]
for user in users:
    profile = Profile.from_username(L.context, user)
    for post in profile.get_posts():
        L.download_post(post, target=user)
```

```python
L = Instaloader()
for post in L.get_hashtag_posts(HASHTAG):
    L.download_post(post, target='#'+HASHTAG)
```

Unzip metadata files

```python
for user in users:
    source = user + '/'
    destination = user + '_metadata/'
    if not os.path.exists(destination):
        os.mkdir(destination)
        print("Directory " , destination ,  " Created ")
    else:    
        print("Directory " , destination ,  " already exists")
        
    for element in os.listdir(source):
        if element.endswith('.json.xz'):
            print("'%s' will move to destination" % element)
            shutil.move(source + element, destination)
        
```

```python
for user in users:
    directory = user + '_metadata/'
    destination = directory + 'metadata_json/'
    if not os.path.exists(destination):
        os.mkdir(destination)
        print("Directory " , destination ,  " Created ")
    else:    
        print("Directory " , destination ,  " already exists")
    for element in os.listdir(directory):
        if element.endswith('.json.xz'):
            rename_ele = element[:-2]
            file = open(destination + rename_ele, "w")
            file.write(lzma.open(directory + element).read().decode("utf-8") )
            file.close()
```

Select data that matter

```python
user_name = []
location = []
caption = []
i = 1
for user in users:
    directory = user + '_metadata/metadata_json/'
    for element in os.listdir(directory):
        user_name.append(user)
        with open(directory + element) as json_file:
            post = json.load(json_file)
            print('Photo n°: {}'.format(i))
            if post['node'] is not None:
                if post['node']['location'] is not None:
                    print("Location Name : ", post['node']['location']['name'])
                    print("Location Slug : ", post['node']['location']['slug'])
                    location.append(post['node']['location']['slug'])
                else:
                    print("Location Name : -\nLocation Slug : -")
                    location.append("")
                if 'accessibility_caption' in post['node']:
                    print("Caption : ", post['node']['accessibility_caption'])
                    caption.append(post['node']['accessibility_caption'])
                else:
                    print("Caption : -")
                    caption.append("")
            i += 1
```

```python
len(user_name), len(location), len(caption)
```

```python
dataset = {'User_Name': user_name, 'Location': location, 'Caption': caption}
df = pd.DataFrame(dataset)
df
```

```python
string_empty = df[ (df['Location'] == '') & (df['Caption'] == '') ].index
df.drop(string_empty , inplace=True)

no_description = df[ (df['Location'] == '') & (df['Caption'] == 'No photo description available.') ].index
df.drop(no_description , inplace=True)
```

```python
df_location = df.copy() # Will be used to determine which is a traveler
df_description = df.copy() # will be used to determine wich is a foody / doggy person 
```

```python
no_location = df_location[(df_location['Location'] == '')].index
df_location.drop(no_location , inplace=True)

by_user_by_location = pd.DataFrame(df_location.groupby(['User_Name'])['Location'].nunique())
by_user_by_location
```

```python
def is_traveler(row):
    if row['Location'] >= 10 :
        label = True
    else:
        label = False
    return label
```

```python
by_user_by_location['traveler'] = by_user_by_location.apply(is_traveler, axis=1)
```

```python
no_description = df_description[(df_description['Caption'] == '')].index
df_description.drop(no_description , inplace=True)
no_description = df_description[(df_description['Caption'] == 'No photo description available.')].index
df_description.drop(no_description , inplace=True)
df_description
```

```python
description_by_user = [] # Liste de dataframe
for user in users:
    is_user =  df_description['User_Name'] == user
    description_by_user.append(pd.DataFrame(df_description[is_user]))
```

```python
text_concat = df_description.groupby('User_Name').apply(lambda x: ' '.join(x.Caption))
text_concat
```

```python
count = CountVectorizer()
bag_of_words = count.fit_transform(text_concat)

bag_of_words.toarray()
```

```python
feature_names = count.get_feature_names()
feature_names
```

```python
bow_description = pd.DataFrame(bag_of_words.toarray(), columns=feature_names)
bow_description
```

```python
def is_foody(row):
    label = False
    if row['coffee'] >= 1 :
        label = True
    elif row['drink'] >= 100:
        label = True
    elif row['eating'] >= 100:
        label = True
    elif row['food'] >= 100:
        label = True
    elif row['pizza'] >= 10:
        label = True
    return label
```

```python
col_foody = bow_description.apply(is_foody, axis=1)
col_foody
```

```python
df_text = pd.DataFrame(text_concat)
df_text
```

```python
values_foody = []
for val in col_foody:
    values_foody.append(val)
```

```python
df_text['foody'] = values_foody
```
